```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import random
from typing import Tuple, List
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

class ManifoldEnvironment:
    def __init__(self, manifold_type: str = 'sphere', params: dict = None):
        self.manifold_type = manifold_type
        self.params = params or self._default_params()
        self.current_position = self._random_point()
        self.trajectory = [self.current_position]
        self.frame = self._initial_frame()
    
    def _default_params(self) -> dict:
        defaults = {
            'sphere': {'radius': 2.0},
            'torus': {'R': 3.0, 'r': 1.0},
            'hyperbolic': {'k': -1.0}
        }
        return defaults[self.manifold_type]
    
    def _random_point(self) -> np.ndarray:
        if self.manifold_type == 'sphere':
            theta = np.random.uniform(0, 2 * np.pi)
            phi = np.arccos(np.random.uniform(-1, 1))
            return self.params['radius'] * np.array([
                np.sin(phi) * np.cos(theta),
                np.sin(phi) * np.sin(theta),
                np.cos(phi)
            ])
        
        elif self.manifold_type == 'torus':
            u = np.random.uniform(0, 2 * np.pi)
            v = np.random.uniform(0, 2 * np.pi)
            R, r = self.params['R'], self.params['r']
            return np.array([
                (R + r * np.cos(v)) * np.cos(u),
                (R + r * np.cos(v)) * np.sin(u),
                r * np.sin(v)
            ])
        
        elif self.manifold_type == 'hyperbolic':
            r = np.random.uniform(0, 0.9)
            theta = np.random.uniform(0, 2 * np.pi)
            return np.array([r * np.cos(theta), r * np.sin(theta), 0])
    
    def _initial_frame(self) -> np.ndarray:
        if self.manifold_type == 'sphere':
            pos = self.current_position
            r = self.params['radius']
            theta = np.arctan2(pos[1], pos[0])
            phi = np.arccos(pos[2] / r)
            
            e1 = np.array([np.cos(theta) * np.cos(phi),
                          np.sin(theta) * np.cos(phi),
                          -np.sin(phi)])
            e2 = np.array([-np.sin(theta),
                          np.cos(theta),
                          0])
            return np.stack([e1, e2])
        
        elif self.manifold_type == 'torus':
            pos = self.current_position
            R, r = self.params['R'], self.params['r']
            u = np.arctan2(pos[1], pos[0])
            v = np.arctan2(pos[2], np.sqrt(pos[0]**2 + pos[1]**2) - R)
            
            e1 = np.array([-np.sin(u), np.cos(u), 0])
            e2 = np.array([np.cos(u) * (-np.sin(v)),
                          np.sin(u) * (-np.sin(v)),
                          np.cos(v)])
            return np.stack([e1, e2])
        
        elif self.manifold_type == 'hyperbolic':
            pos = self.current_position[:2]
            theta = np.arctan2(pos[1], pos[0])
            e1 = np.array([np.cos(theta), np.sin(theta), 0])
            e2 = np.array([-np.sin(theta), np.cos(theta), 0])
            scale = 1 / (1 - np.sum(pos**2))
            return scale * np.stack([e1, e2])
    
    def parallel_transport(self, frame: np.ndarray, displacement: np.ndarray) -> np.ndarray:
        if self.manifold_type == 'sphere':
            new_pos = self.current_position + displacement
            new_pos = new_pos / np.linalg.norm(new_pos) * self.params['radius']
            normal = new_pos / np.linalg.norm(new_pos)
            
            new_frame = []
            for vec in frame:
                transported = vec - np.dot(vec, normal) * normal
                transported = transported / np.linalg.norm(transported)
                new_frame.append(transported)
            return np.stack(new_frame)
            
        elif self.manifold_type == 'torus':
            new_pos = self.current_position + displacement
            R, r = self.params['R'], self.params['r']
            u = np.arctan2(new_pos[1], new_pos[0])
            d = np.sqrt(new_pos[0]**2 + new_pos[1]**2)
            v = np.arctan2(new_pos[2], d - R)
            normal = np.array([np.cos(u) * np.cos(v),
                            np.sin(u) * np.cos(v),
                            np.sin(v)])
            
            new_frame = []
            for vec in frame:
                transported = vec - np.dot(vec, normal) * normal
                transported = transported / np.linalg.norm(transported)
                new_frame.append(transported)
            return np.stack(new_frame)
            
        elif self.manifold_type == 'hyperbolic':
            new_pos = self.current_position + displacement
            scale = 1 / (1 - np.sum(new_pos[:2]**2))
            new_frame = []
            for vec in frame:
                transported = scale * vec
                if np.linalg.norm(transported) > 0:
                    transported = transported / np.linalg.norm(transported)
                new_frame.append(transported)
            return np.stack(new_frame)
        
        elif self.manifold_type == 'torus':
            new_pos = self.current_position + displacement
            R, r = self.params['R'], self.params['r']
            u = np.arctan2(new_pos[1], new_pos[0])
            v = np.arctan2(new_pos[2], np.sqrt(new_pos[0]**2 + new_pos[1]**2) - R)
            normal = np.array([np.cos(u) * np.cos(v),
                             np.sin(u) * np.cos(v),
                             np.sin(v)])
            return vector - np.dot(vector, normal) * normal
        
        elif self.manifold_type == 'hyperbolic':
            new_pos = self.current_position + displacement
            scale = 1 / (1 - np.sum(new_pos[:2]**2))
            return scale * vector
    
    def gaussian_curvature(self, point: np.ndarray) -> float:
        if self.manifold_type == 'sphere':
            return 1.0 / (self.params['radius'] ** 2)
        
        elif self.manifold_type == 'torus':
            R, r = self.params['R'], self.params['r']
            u = np.arctan2(point[1], point[0])
            v = np.arctan2(point[2], np.sqrt(point[0]**2 + point[1]**2) - R)
            return np.cos(v) / (r * (R + r * np.cos(v)))
        
        elif self.manifold_type == 'hyperbolic':
            return self.params['k']
    
    def visualize_trajectory(self):
        trajectory = np.array(self.trajectory)
        fig = plt.figure(figsize=(15, 5))
        
        if self.manifold_type in ['sphere', 'torus']:
            ax = fig.add_subplot(121, projection='3d')
            
            # Set frame scale based on manifold
            if self.manifold_type == 'sphere':
                frame_scale = 0.5
                u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
                r = self.params['radius']
                x = r * np.cos(u) * np.sin(v)
                y = r * np.sin(u) * np.sin(v)
                z = r * np.cos(v)
                ax.plot_wireframe(x, y, z, color='gray', alpha=0.2)
            else:  # torus
                frame_scale = 0.3
                u, v = np.mgrid[0:2*np.pi:30j, 0:2*np.pi:20j]
                R, r = self.params['R'], self.params['r']
                x = (R + r * np.cos(v)) * np.cos(u)
                y = (R + r * np.cos(v)) * np.sin(u)
                z = r * np.sin(v)
                ax.plot_wireframe(x, y, z, color='gray', alpha=0.2)
            
            # Plot trajectory and frame
            ax.plot(trajectory[:, 0], trajectory[:, 1], trajectory[:, 2], 'r-', 
                   label='Path', linewidth=2)
            ax.scatter(trajectory[-1, 0], trajectory[-1, 1], trajectory[-1, 2], 
                      color='red', s=100, label='Current')
            
            pos = self.current_position
            frame = self.frame
            ax.quiver(pos[0], pos[1], pos[2], 
                     frame[0, 0], frame[0, 1], frame[0, 2],
                     color='blue', length=frame_scale, label='e1')
            ax.quiver(pos[0], pos[1], pos[2],
                     frame[1, 0], frame[1, 1], frame[1, 2],
                     color='green', length=frame_scale, label='e2')
            
        else:  # Hyperbolic plane
            ax = fig.add_subplot(121)
            circle = plt.Circle((0, 0), 1, fill=False, color='gray')
            ax.add_artist(circle)
            ax.plot(trajectory[:, 0], trajectory[:, 1], 'r-', 
                   label='Path', linewidth=2)
            ax.scatter(trajectory[-1, 0], trajectory[-1, 1],
                      color='red', s=100, label='Current')
            
            pos = self.current_position
            frame = self.frame
            r = np.linalg.norm(pos[:2])
            frame_scale = 0.2 * (1 - r)  # Scale with distance from boundary
            
            ax.quiver(pos[0], pos[1], frame[0, 0], frame[0, 1],
                     color='blue', scale=1/frame_scale, label='e1')
            ax.quiver(pos[0], pos[1], frame[1, 0], frame[1, 1],
                     color='green', scale=1/frame_scale, label='e2')
            ax.set_xlim(-1.1, 1.1)
            ax.set_ylim(-1.1, 1.1)
            ax.set_aspect('equal')
        
        ax.set_title(f'{self.manifold_type.capitalize()} Manifold')
        ax.legend()
        
        # Plot curvature
        ax2 = fig.add_subplot(122)
        curvatures = [self.gaussian_curvature(pos) for pos in trajectory]
        ax2.plot(curvatures, linewidth=2)
        ax2.set_title('Gaussian Curvature along Path')
        ax2.set_xlabel('Step')
        ax2.set_ylabel('Curvature')
        
        plt.tight_layout()
        plt.show()

class GeometricAgent(nn.Module):
    def __init__(self, state_dim: int = 4, action_dim: int = 3, hidden_dim: int = 64):
        super().__init__()
        self.embedding = nn.Sequential(
            nn.Linear(state_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU()
        )
        self.policy_head = nn.Linear(hidden_dim, action_dim)
        self.value_head = nn.Linear(hidden_dim, 1)
    
    def forward(self, state: torch.Tensor) -> Tuple[torch.Tensor, torch.Tensor]:
        embedding = self.embedding(state)
        action_probs = torch.tanh(self.policy_head(embedding))
        state_value = self.value_head(embedding)
        return action_probs, state_value

class GeometricLearner:
    def __init__(self, env: ManifoldEnvironment, agent: GeometricAgent, learning_rate: float = 3e-4):
        self.env = env
        self.agent = agent
        self.optimizer = torch.optim.Adam(agent.parameters(), lr=learning_rate)
        self.replay_buffer = []
        self.buffer_size = 1000
        self._last_positions = np.zeros((10, 3))
    
    def compute_state_representation(self, position: np.ndarray) -> torch.Tensor:
        curvature = self.env.gaussian_curvature(position)
        state = np.concatenate([position, [curvature]])
        return torch.FloatTensor(state).unsqueeze(0).float()
    
    def _compute_reward(self, old_pos: np.ndarray, new_pos: np.ndarray) -> float:
        distance_moved = np.linalg.norm(new_pos - old_pos)
        
        if self.env.manifold_type == 'sphere':
            exploration_reward = distance_moved
        elif self.env.manifold_type == 'torus':
            # Reward exploring major radius
            u_old = np.arctan2(old_pos[1], old_pos[0])
            u_new = np.arctan2(new_pos[1], new_pos[0])
            major_circle_progress = abs(u_new - u_old)
            exploration_reward = distance_moved + 0.5 * major_circle_progress
        else:  # hyperbolic
            # Reward approaching boundary
            r = np.linalg.norm(new_pos[:2])
            boundary_reward = r / (1 - r)
            exploration_reward = distance_moved + 0.3 * boundary_reward
        
        return exploration_reward
    
    def train_step(self) -> dict:
        state = self.compute_state_representation(self.env.current_position)
        action_probs, _ = self.agent(state)
        
        # Adjust exploration noise by manifold
        if self.env.manifold_type == 'sphere':
            noise = 0.2
        elif self.env.manifold_type == 'torus':
            noise = 0.3  # More exploration for torus
        else:  # hyperbolic
            r = np.linalg.norm(self.env.current_position[:2])
            noise = 0.2 * (1 + r)  # More exploration near boundary
        
        action_distribution = torch.distributions.Normal(action_probs.squeeze(), noise)
        action = action_distribution.sample()
        log_prob = action_distribution.log_prob(action).sum()
        
        action_np = action.detach().numpy()
        
        # Project action to tangent space
        if self.env.manifold_type == 'sphere':
            normal = self.env.current_position / np.linalg.norm(self.env.current_position)
            action_np = action_np - np.dot(action_np, normal) * normal
            action_np = action_np / np.linalg.norm(action_np)
        
        # Adjust step size by manifold
        if self.env.manifold_type == 'sphere':
            step_size = 0.1
        elif self.env.manifold_type == 'torus':
            step_size = 0.2
        else:  # hyperbolic
            r = np.linalg.norm(self.env.current_position[:2])
            step_size = 0.1 / (1 - r)
        
        new_position = self.env.current_position + step_size * action_np
        
        # Project to manifold
        if self.env.manifold_type == 'sphere':
            new_position = new_position * self.env.params['radius'] / np.linalg.norm(new_position)
        elif self.env.manifold_type == 'torus':
            R, r = self.env.params['R'], self.env.params['r']
            u = np.arctan2(new_position[1], new_position[0])
            d = np.sqrt(new_position[0]**2 + new_position[1]**2)
            v = np.arctan2(new_position[2], d - R)
            new_position = np.array([
                (R + r * np.cos(v)) * np.cos(u),
                (R + r * np.cos(v)) * np.sin(u),
                r * np.sin(v)
            ])
        elif self.env.manifold_type == 'hyperbolic':
            if np.linalg.norm(new_position[:2]) >= 1:
                new_position[:2] = new_position[:2] / np.linalg.norm(new_position[:2]) * 0.99
            new_position[2] = 0
        
        reward = self._compute_reward(self.env.current_position, new_position)
        next_state = self.compute_state_representation(new_position)
        
        self.env.trajectory.append(new_position)
        self.env.frame = self.env.parallel_transport(self.env.frame, new_position - self.env.current_position)
        self.env.current_position = new_position
        
        return {
            'position': new_position,
            'reward': reward,
            'action': action_np
        }

manifolds = ['sphere', 'torus', 'hyperbolic']
for manifold_type in manifolds:
    print(f"\nTraining on {manifold_type.capitalize()} manifold:")
    
    # Create environment with appropriate parameters
    params = None
    if manifold_type == 'sphere':
        params = {'radius': 2.0}
    elif manifold_type == 'torus':
        params = {'R': 3.0, 'r': 1.0}
    elif manifold_type == 'hyperbolic':
        params = {'k': -1.0}
    
    env = ManifoldEnvironment(manifold_type=manifold_type, params=params)
    agent = GeometricAgent(state_dim=4, action_dim=3)
    learner = GeometricLearner(env, agent)
    
    rewards = []
    for episode in range(50):
        step_info = learner.train_step()
        rewards.append(step_info['reward'])
        
        if episode % 10 == 0:
            print(f"Episode {episode}: Reward = {step_info['reward']:.3f}")
            env.visualize_trajectory()
            
    plt.figure(figsize=(8, 4))
    plt.plot(rewards)
    plt.title(f'{manifold_type.capitalize()} Manifold Learning Curve')
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.show()
```
